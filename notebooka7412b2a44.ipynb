```python
# text from image
from tempfile import NamedTemporaryFile
import requests
def text_from_image(image_link):
    try:
        with NamedTemporaryFile(delete=False, suffix=".jpg") as temp_file:
            # Download the image
            response = requests.get(image_link)
            response.raise_for_status()  # Ensure we got a successful response
            temp_file.write(response.content)
            temp_image_path = temp_file.name
        
        # Perform OCR
        results = reader.readtext(temp_image_path, detail=0)
        
        # Remove the temporary file
        os.remove(temp_image_path)
        
        # Return the extracted text
        if results:
            return " ".join(results)
        else:
            return ""
    
    except Exception as e:
        print(f"[ERROR] Failed to extract text: {e}")
        return ""
```

```python
sorted_data = data.sort_values(by='group_id')
```

```python
sorted_data.head(5)
```

```python
previous_group_id = None
previous_text = ""
```

```python
def process_row(row):
    global previous_group_id, previous_text
    # Check if the current group_id is the same as the previous one
    if row['group_id'] == previous_group_id:
        return previous_text  # Reuse the previous extracted text

    # If different, perform OCR
    extracted_text = text_from_image(row['image_link'])
    # Update the previous group_id and text
    previous_group_id = row['group_id']
    previous_text = extracted_text
    return extracted_text
```

```python
tqdm.pandas(desc="Processing images")
sorted_data['extracted_text'] = sorted_data.progress_apply(process_row, axis=1)
```

```python
output_file = 'extracted_texts_sample_sorted.csv'  # Save to a persistent location
sorted_data.to_csv(output_file, index=False)
```
